```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb['neighbourhood_group'].value_counts()
#There's 5 neighbourhood groups, Manhattan shows up the most
```

```python
# Are private rooms the most popular in manhattan?
air_bnb['room_type'].value_counts()
#No, Entire home/apt are the most popular
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.groupby(['host_name'])['number_of_reviews'].sum().sort_values(ascending = False).head()

#it looks like Michael has the most reviews
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby('neighbourhood_group').mean()
#Staten Island has the highest Average Price
```

```python
# Which neighborhood group has the highest total price?
airbnb=air_bnb.sort_values(by=['price'],ascending=False).head(100)
airbnb['neighbourhood_group'].value_counts()
#Manhattan has the highest total price
```

```python
#Which top 5 hosts have the highest total price?

air_bnb.groupby(['host_id', 'host_name'])['price'].sum().sort_values(ascending = False).head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb['reviews_per_month'] < 100
```

```python
# What host has the highest total of prices and where are they located?

air_bnb.price.describe()
```

```python
air_bnb[air_bnb['price'] == 10000.00]
#Kathrine has the highest price and they're in Queens, bc min nights are @100
```

```python
# When did Danielle from Queens last receive a review?
air_bnb[(air_bnb.host_name == 'Danielle') & (air_bnb.neighbourhood_group == 'Queens')].groupby(['last_review']).max()
#It looks like there are multiple Danielle's loool
```
